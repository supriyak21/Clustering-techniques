Task4. Find out why we need to maintain a uniform scale across the variables for K-means and Hierarchical clustering

As distance measures are used by both K-means and Hierarchical clustering methods to evaluate how similar two data points are, maintaining a consistent scale across variables is essential.Following are the reasons, why we need to maintain a uniform scale across the variables for K-means and Hierarchial Clustering:

1.Considering every variable equally : When scaling is not used, the influence of factors with smaller magnitudes (like age in years) is largely overshadowed by variables with greater magnitudes (like income in dollars). Clusters that prioritise one dimension over others are produced as a result of this bias.In order to ensure that every variable contributes equally to the distance computation and to produce more meaningful clusters based on actual correlations in the data, scaling levels the playing field for all variables.

2.Smooth convergence : K-means distance computations depend on assigning data points and updating centroids iteratively. Because large-magnitude variables predominate when scales are uneven, the algorithm becomes trapped in local optima. This results in potentially less-than-ideal clustering outcomes and slower convergence. Scaling facilitates the algorithm's more efficient exploration of the whole data space, which could result in faster convergence and better answers.

3.Interpretation that is consistently true: Cluster interpretation becomes challenging when variables have distinct dimensions.The clustering results can be coherently interpreted using scaling, which enables meaningful comparisons of cluster attributes and distances across all variables.

Task6. Implement K-means clustering to categorize the data into clusters 

```python
# K Means clustering(By taking 3 clusters)

import random
# Choose the number of clusters (e.g., k = 3)
k = 3
#distance between two points is calculated by Euclidean distance
def distance(point1, point2):
   total_distance = 0
   for i in range(len(point1)):
    distance_squared = (point1[i] - point2[i])**2
    total_distance += distance_squared
   return total_distance**0.5  # Square root for final distance
# Assign each datapoint to nearest centroid
def assign_to_clusters(data, centroids):
# Creating empty clusters 
  clusters = [[] for _ in range(k)]  
  for index, row in numerical_cols.iterrows():  # Iterate through DataFrame rows
# Convert DataFrame row to a list      
    point = row.tolist()  
    closest_centroid_index = 0
    closest_distance = float('inf')  # Set to infinity initially
    for i, centroid in enumerate(centroids):
      current_distance = distance(point, centroid)
      if current_distance < closest_distance:
        closest_distance = current_distance
        closest_centroid_index = i
    clusters[closest_centroid_index].append(index)  # Append row index to cluster
  return clusters

def update_centroids(clusters, data):
#Since, centroids are the mean of the datapoints in the cluster, therefore updating the centroids to be the averages of their assigned data points
  new_centroids = []
  for cluster in clusters:
#empty clusters (no datapoints)      
    if not cluster:  
      new_centroids.append(data.iloc[random.randint(0, len(data)-1)].tolist())  # Replace with random point
    else:
      new_centroid = [0] * len(data.columns)  # Initialize new centroid with zeros
      for index in cluster:
        row = numerical_cols.iloc[index]  # Access DataFrame row using index
        for i in range(len(numerical_cols.columns)):
          new_centroid[i] += row[i]
      for i in range(len(new_centroid)):
        new_centroid[i] /= len(cluster)  # Calculate average
      new_centroids.append(new_centroid)
  return new_centroids

# Choose random initial centroids from the DataFrame
centroids = numerical_cols.sample(k).values.tolist()  # Convert to lists for consistency

# Loop until centroids don't change significantly
max_iterations = 100
for _ in range(max_iterations):
  new_clusters = assign_to_clusters(numerical_cols, centroids)
  new_centroids = update_centroids(new_clusters, numerical_cols)

# Check if centroids converged (e,e. centrpoids are not changing much)
  converged = True
  for i in range(k):
    for j in range(len(centroids[i])):
      if abs(centroids[i][j] - new_centroids[i][j]) > 0.01:  # Threshold for change
        converged = False
        break
    if not converged:
      break

  if converged:
    break
  else:
    centroids = new_centroids

# Display the final clusters and centroids
print("Clusters:")
for i, cluster in enumerate(new_clusters):
  print(f"Cluster {i+1}:", numerical_cols.iloc[cluster])  # Access DataFrame rows using indices
print("\nCentroids:")
for i, centroid in enumerate(new_centroids):
  print(f"Centroid {i+1}:", centroid)
```

Task7. Use elbow method to determine number of clusters 

```python
from sklearn.cluster import KMeans 
# To store the number of clusters(K), k[] representas a list
k = [] 
# To store the within cluster sum of squares(WCSS), wcss[] represents a list
wcss = []
# creating a loop for i clusters from KMeans model within the range (1,10)
for i in range(1,10):
  mdl = KMeans(n_clusters=i, n_init=10)
  mdl.fit(df.iloc[:,3:10])
  k.append(i)
  wcss.append(mdl.inertia_)
```

```python
# Plotting a line graph with number of clusters on x axis and wcss on y axis.
sns.lineplot(x = k, y = wcss)
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.show()
```

Insights: The point where the rate of decrease in wcss slows down, determines the optimal number of clusters for KMeans algorithm. From the graph, we can visualize that a point 3,the decrease in WCSS slows down. So, the optimal number of clusters for this dataset is 3.

Task8. Implement Hierarchical clustering using an appropriate linkage method.

```python
# Hierarchical clustering

def euclidean_distance(a, b):
    return np.linalg.norm(a - b)

def complete_linkage(cluster1, cluster2):
    max_distance = 0

    for point1 in cluster1:
        for point2 in cluster2:
            distance = euclidean_distance(point1, point2)
            max_distance = max(max_distance, distance)

    return max_distance
#DataFrame (numerical_cols) is subjected to hierarchical clustering.Each data point is first initialised as a cluster, and iteratively, up to the point where only one cluster is left, the two clusters are merged using the shortest total linkage distance. The function returns the linkage matrix including the merging information along with the resultant cluster
def hierarchical_clustering(numerical_cols):
    clusters = [np.array(row) for _, row in numerical_cols.iterrows()]

    linkage_matrix = []

    while len(clusters) > 1:
        min_distance = float('inf')
        merge_indices = (0, 0)

        for i in range(len(clusters)):
            for j in range(i + 1, len(clusters)):
                distance = complete_linkage(clusters[i], clusters[j])

                if distance < min_distance:
                    min_distance = distance
                    merge_indices = (i, j)

        merged_cluster = np.vstack((clusters[merge_indices[0]], clusters[merge_indices[1]]))
        clusters.pop(merge_indices[1])
        clusters[merge_indices[0]] = merged_cluster

        linkage_matrix.append([merge_indices[0], merge_indices[1], min_distance, len(merged_cluster)])

    return clusters[0], np.array(linkage_matrix)
    
#Plotting a dendrogram after receiving the linkage matrix as input using the plot_dendrogram function.
def plot_dendrogram(linkage_matrix):
    plt.figure(figsize=(10, 5))
    plt.title('Hierarchical Clustering Dendrogram')
    plt.xlabel('Cluster Index')
    plt.ylabel('Distance')

    for i in range(len(linkage_matrix)):
        x = [linkage_matrix[i, 0], linkage_matrix[i, 0]]
        y = [linkage_matrix[i, 1], linkage_matrix[i, 1]]
        plt.plot(x, y, color='black')

        x = [linkage_matrix[i, 0], i, i, linkage_matrix[i, 1]]
        y = [linkage_matrix[i, 1], linkage_matrix[i, 1], linkage_matrix[i, 3], linkage_matrix[i, 3]]
        plt.plot(x, y, color='black')

    plt.show()
#The data is prepared, hierarchical clustering is applied, the resultant cluster is printed, and the dendrogram is plotted.
# fitting it with "numerical_cols" dataframe
data = numerical_cols

# Perform hierarchical clustering
resulting_cluster, linkage_matrix = hierarchical_clustering(numerical_cols)

# Display the resulting cluster
print("Resulting Cluster:")
print(resulting_cluster)

# Plot dendrogram
plot_dendrogram(linkage_matrix)
```
